# Polars Bokeh Accessor Examples

Minimal examples using the `pl.DataFrame.bokeh` accessor.

```python
%load_ext autoreload
%autoreload 2

import polars as pl

from bokeh.core.enums import *
from bokeh.colors.groups import *
# from bokeh.colors.named import *
from bokeh.transform import transform
from bokeh.io import show
from bokeh.io import output_notebook
from spectral.charts import BokehAccessor
from spectral.charts.theme_manager import theme

output_notebook(hide_banner=True)
theme.set("light_minimal")
```

```python
theme.set("ocean")
df = pl.DataFrame(
    {"x": [1, 2, 3, 4], "y": [1, 4, 2, 8], "y2": [2, 5, 3, 9]}
)
fig = df.bokeh(title="Ocean scatter styling", width=700, height=300, tools="pan,wheel_zoom,reset")
fig.line(
    x="x",
    y="y",
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=blue[0],
    line_dash=DashPattern.dashdot,
    line_dash_offset=0,
    line_join=LineJoin.miter,
    line_width=2,
)
fig.varea(
    x="x",
    y1="y",
    y2="y2",
    fill_alpha=1,
    fill_color=blue[3],
    hatch_alpha=1,
    hatch_color=brown[0],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=5,
    hatch_weight=1,
)
fig.scatter(
    x="x",
    y="y",
    fill_alpha=1,
    fill_color=blue[1],
    hatch_alpha=1,
    hatch_color=blue[2],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=10.25,
    hatch_weight=1.0,
    hit_dilation=1,
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=blue[6],
    line_dash=DashPattern.solid,
    line_dash_offset=0,
    line_join=LineJoin.round,
    line_width=2,
    marker=MarkerType.cross,
    size=8,
)
show(fig)
```
