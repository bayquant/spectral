# Polars Bokeh Accessor Examples

Minimal examples using the `pl.DataFrame.bokeh` accessor.

```python
%load_ext autoreload
%autoreload 2

import polars as pl

from bokeh.core.enums import *
from bokeh.colors import groups
from bokeh.colors import named
from bokeh.transform import transform
from bokeh.io import show
from bokeh.io import output_notebook
from spectral.charts import BokehAccessor
from spectral.charts.theme_manager import theme

output_notebook(hide_banner=True)
theme.set("ocean")
```

```python
df = pl.DataFrame(
    {"x": [1, 2, 3, 4], "y": [1, 4, 2, 8], "y2": [2, 5, 3, 9]}
)
fig = df.bokeh(title="Ocean scatter styling", width=700, height=300, tools="pan,wheel_zoom,reset")
fig.line(
    x="x",
    y="y",
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=groups.blue[0],
    line_dash=DashPattern.dashdot,
    line_dash_offset=0,
    line_join=LineJoin.miter,
    line_width=2,
    legend_label="T"
)
fig.varea(
    x="x",
    y1="y",
    y2="y2",
    fill_alpha=1,
    fill_color=groups.blue[3],
    hatch_alpha=1,
    hatch_color=groups.brown[0],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=5,
    hatch_weight=1,
    legend_label="T1"
)
fig.scatter(
    x="x",
    y="y",
    fill_alpha=1,
    fill_color=groups.blue[1],
    hatch_alpha=1,
    hatch_color=groups.blue[2],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=10.25,
    hatch_weight=1.0,
    hit_dilation=1,
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=groups.blue[6],
    line_dash=DashPattern.solid,
    line_dash_offset=0,
    line_join=LineJoin.round,
    line_width=2,
    marker=MarkerType.cross,
    size=8,
    legend_label="T2"
)
fig.legend.location = "top_left"
fig.legend.click_policy = "hide"
show(fig)
```

```python
from bokeh.layouts import row, column
from bokeh.plotting import figure, show

def make_df() -> pl.DataFrame:
    return pl.DataFrame(
        {
            "x": [1, 2, 3, 4],
            "y1": [1, 4, 2, 3],
            "y2": [4, 3, 2, 1],
            "y3": [2, 5, 3, 6],
            "y4": [2, 1, 4, 3],
        }
    )

p1 = make_df().bokeh(title="Plot 1", width=450, height=300)
p1.line(x="x", y="y1", line_width=2, line_color=groups.blue[0])

p2 = make_df().bokeh(title="Plot 2", width=450, height=300)
p2.scatter(x="x", y="y2", size=10, color=groups.brown[5], line_color=groups.brown[5])

p3 = make_df().bokeh(title="Plot 3", width=450, height=300)
p3.vbar(x="x", top="y3", width=0.5, fill_color=groups.cyan[0], line_color=groups.cyan[0])

p4 = make_df().bokeh(title="Plot 4", width=450, height=300)
p4.scatter(x="x", y="y4", size=12, color=groups.purple[4], line_color=groups.purple[4])

layout = column(
    row(p1, p2, spacing=10),
    row(p3, p4, spacing=10),
    spacing=10
)

show(layout)
```

```python
df = pl.DataFrame(
    {
        "x": [1, 2, 3, 4],
        "series_1": [1, 4, 2, 3],
        "series_2": [2, 1, 4, 5],
        "series_3": [3, 5, 1, 2],
    }
)

fig = df.bokeh(title="Three labeled lines", width=700, height=300, tools="pan,wheel_zoom,reset")
fig.line(
    x="x",
    y="series_1",
    line_width=2,
    line_color=groups.blue[0],
    legend_label="Series 1",
)
fig.line(
    x="x",
    y="series_2",
    line_width=2,
    line_color=groups.green[0],
    legend_label="Series 2",
)
fig.line(
    x="x",
    y="series_3",
    line_width=2,
    line_color=groups.red[0],
    legend_label="Series 3",
)

fig.legend.location = "top_left"
fig.legend.click_policy = "hide"
show(fig)
```
