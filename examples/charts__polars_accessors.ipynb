# Polars Bokeh Accessor Examples

Minimal examples using the `pl.DataFrame.bokeh` accessor.

```python
%load_ext autoreload
%autoreload 2

# Other imports
from bokeh.colors import groups
from bokeh.colors import named
from bokeh.core.enums import *
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.layouts import column
from bokeh.layouts import row
from bokeh.plotting import figure
from bokeh.plotting._figure import FigureOptions
from bokeh.transform import transform
import polars as pl
from xpectral.charts import BokehAccessor
from xpectral.charts.theme_manager import theme

output_notebook(hide_banner=True)
theme.set("ocean")

df = pl.DataFrame({
    "x": [1, 2, 3, 4],
    "y1": [1, 4, 2, 3],
    "y2": [2, 5, 3, 4],
    "y3": [4, 2, 3, 1],
    "y4": [3, 1, 4, 2],
    "y5": [1, 2, 1, 2],
    "y6": [1, 1, 2, 1],
})
```

```python
fig = df.bokeh(title="Ocean scatter styling", width=700, height=300, tools="pan,wheel_zoom,reset")
fig.line(
    x="x",
    y="y1",
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=groups.blue[0],
    line_dash=DashPattern.dashdot,
    line_dash_offset=0,
    line_join=LineJoin.miter,
    line_width=2,
    legend_label="y",
)
fig.varea(
    x="x",
    y1="y1",
    y2="y2",
    fill_alpha=1,
    fill_color=groups.blue[3],
    hatch_alpha=1,
    hatch_color=groups.brown[0],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=5,
    hatch_weight=1,
)
fig.scatter(
    x="x",
    y="y1",
    fill_alpha=1,
    fill_color=groups.blue[1],
    hatch_alpha=1,
    hatch_color=groups.blue[2],
    hatch_pattern=HatchPattern.cross,
    hatch_scale=10.25,
    hatch_weight=1.0,
    hit_dilation=1,
    line_alpha=1,
    line_cap=LineCap.round,
    line_color=groups.blue[6],
    line_dash=DashPattern.solid,
    line_dash_offset=0,
    line_join=LineJoin.round,
    line_width=2,
    marker=MarkerType.cross,
    size=8,
)
fig.legend.location = "top_left"
fig.legend.click_policy = "hide"
show(fig)
```

```python
p1 = df.bokeh(title="Plot 1", width=450, height=300)
p1.line(x="x", y="y1", line_width=2, line_color=groups.blue[0])

p2 = df.bokeh(title="Plot 2", width=450, height=300)
p2.scatter(x="x", y="y2", size=10, color=groups.brown[5], line_color=groups.brown[5])

p3 = df.bokeh(title="Plot 3", width=450, height=300)
p3.vbar(x="x", top="y3", width=0.5, fill_color=groups.cyan[0], line_color=groups.cyan[0])

p4 = df.bokeh(title="Plot 4", width=450, height=300)
p4.scatter(x="x", y="y4", size=12, color=groups.purple[4], line_color=groups.purple[4])

layout = column(
    row(p1, p2, spacing=10),
    row(p3, p4, spacing=10),
    spacing=10
)

show(layout)
```

```python
p1 = df.bokeh(title="harea_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p1.harea_stack(["y5", "y6"], y="x", fill_color=[groups.blue[2], groups.blue[6]], fill_alpha=[0.6, 0.9])

p2 = df.bokeh(title="hbar_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p2.hbar_stack(["y5", "y6"], y="x", height=0.5, fill_color=[groups.green[2], groups.green[6]], fill_alpha=[0.6, 0.9], line_color="white", line_width=0.5)

p3 = df.bokeh(title="hline_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p3.hline_stack(["y5", "y6"], y="x", line_width=2.5, line_color=[groups.purple[2], groups.purple[6]], line_alpha=[0.7, 1.0])

p4 = df.bokeh(title="varea_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p4.varea_stack(["y5", "y6"], x="x", fill_color=[groups.blue[2], groups.blue[6]], fill_alpha=[0.6, 0.9])

p5 = df.bokeh(title="vbar_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p5.vbar_stack(["y5", "y6"], x="x", width=0.5, fill_color=[groups.green[2], groups.green[6]], fill_alpha=[0.6, 0.9], line_color="white", line_width=0.5)

p6 = df.bokeh(title="vline_stack", width=300, height=250, tooltips=[("x", "@x"), ("y5", "@y5"), ("y6", "@y6")])
p6.vline_stack(["y5", "y6"], x="x", line_width=2.5, line_color=[groups.purple[2], groups.purple[6]], line_alpha=[0.7, 1.0])

layout = column(
    row(p1, p2, p3, spacing=10),
    row(p4, p5, p6, spacing=10),
    spacing=10,
)
show(layout)
```
