```python
from bokeh.plotting._decorators import glyph_method
from bokeh.models.plots import Plot
from bokeh.models.coordinates import CoordinateMapping
from bokeh.models.renderers import GlyphRenderer
from bokeh.models import glyphs
from typing import Any

from bokeh.util.options import Options


class MyAPI:

    @property
    def plot(self) -> Plot | None:
        return self._parent

    @property
    def coordinates(self) -> CoordinateMapping | None:
        return self._coordinates

    def __init__(self, parent: Plot | None = None, coordinates: CoordinateMapping | None = None) -> None:
        self._parent = parent
        self._coordinates = coordinates

    @glyph_method(glyphs.Annulus)
    def annulus(self, **kwargs: Any) -> GlyphRenderer:
        pass
```

```python
from bokeh.core.properties import String, Int

class MyOptions(Options):
    title = String(default="Untitled")
    width = Int(default=400)

kw = {"title": "Hello", "width": 500, "unknown": "keep me"}

opts = MyOptions(kw)   # pops recognized keys out of kw

print(opts.title)      # Hello
print(opts.width)      # 500
print(kw)              # {'unknown': 'keep me'}
```

```python
class figure(Plot, GlyphAPI):
    def __init__(self, *arg, **kw) -> None:
        super().__init__(*arg, **kw)
    
    @property
    def plot(self):
        return self

    @property
    def coordinates(self):
        return None
```
