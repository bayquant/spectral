# Bokeh Axis and Formatter Reference

This notebook shows small, runnable examples for each concrete axis and tick formatter class available in Bokeh `3.8.x`.

It is intended as a design reference for integrating axis/formatter options into `spectral`.

```python
from datetime import datetime, timedelta

from bokeh.io import output_notebook, show
from bokeh.layouts import column
from bokeh.models import (
    BasicTickFormatter,
    CategoricalTickFormatter,
    CustomJSTickFormatter,
    DatetimeTickFormatter,
    LogTickFormatter,
    MercatorTickFormatter,
    NumeralTickFormatter,
    PrintfTickFormatter,
    TimedeltaTickFormatter,
)
from bokeh.plotting import figure

output_notebook()
```

## Axis Examples

```python
# LinearAxis
p_linear = figure(width=520, height=220, title="LinearAxis")
p_linear.line([0, 1, 2, 3], [0, 1, 4, 9], line_width=2)
show(p_linear)
```

```python
# LogAxis
p_log = figure(width=520, height=220, title="LogAxis", y_axis_type="log")
p_log.line([1, 2, 3, 4, 5], [1, 10, 100, 1000, 10000], line_width=2)
show(p_log)
```

```python
# DatetimeAxis
start = datetime(2026, 1, 1)
dates = [start + timedelta(days=i) for i in range(7)]
values = [10, 12, 9, 14, 13, 15, 16]

p_dt = figure(width=520, height=220, title="DatetimeAxis", x_axis_type="datetime")
p_dt.line(dates, values, line_width=2)
show(p_dt)
```

```python
# CategoricalAxis
cats = ["A", "B", "C", "D"]
vals = [4, 7, 3, 6]

p_cat = figure(width=520, height=220, title="CategoricalAxis", x_range=cats)
p_cat.vbar(x=cats, top=vals, width=0.6)
show(p_cat)
```

```python
# MercatorAxis (Web Mercator coordinates)
x_merc = [-13630000, -13620000, -13610000, -13600000]
y_merc = [4540000, 4550000, 4560000, 4570000]

p_merc = figure(
    width=520,
    height=220,
    title="MercatorAxis",
    x_axis_type="mercator",
    y_axis_type="mercator",
)
p_merc.line(x_merc, y_merc, line_width=2)
show(p_merc)
```

```python
# TimedeltaAxis
# Values are in milliseconds by default for timedelta axes.
dur_ms = [0, 5_000, 30_000, 120_000, 300_000]
metric = [1, 3, 2, 6, 4]

p_tdelta = figure(width=520, height=220, title="TimedeltaAxis", x_axis_type="timedelta")
p_tdelta.line(dur_ms, metric, line_width=2)
show(p_tdelta)
```

`Axis` and `ContinuousAxis` are base classes. In practice you usually configure concrete axis types (`linear`, `log`, `datetime`, `mercator`, `timedelta`, `categorical`) through figure arguments or by editing `p.xaxis`/`p.yaxis`.

## Tick Formatter Examples

```python
# BasicTickFormatter
p_basic = figure(width=520, height=220, title="BasicTickFormatter")
p_basic.line([0, 1, 2, 3], [1000, 2000, 3500, 5000], line_width=2)
p_basic.yaxis.formatter = BasicTickFormatter(precision=1)
show(p_basic)
```

```python
# CategoricalTickFormatter
p_cat_fmt = figure(width=520, height=220, title="CategoricalTickFormatter", x_range=["Q1", "Q2", "Q3", "Q4"])
p_cat_fmt.vbar(x=["Q1", "Q2", "Q3", "Q4"], top=[12, 17, 14, 19], width=0.6)
p_cat_fmt.xaxis.formatter = CategoricalTickFormatter()
show(p_cat_fmt)
```

```python
# CustomJSTickFormatter
p_js = figure(width=520, height=220, title="CustomJSTickFormatter")
p_js.line([0, 1, 2, 3], [1_200_000, 2_300_000, 3_100_000, 4_600_000], line_width=2)
p_js.yaxis.formatter = CustomJSTickFormatter(code="return '$' + (tick / 1_000_000).toFixed(1) + 'M'")
show(p_js)
```

```python
# DatetimeTickFormatter
p_dt_fmt = figure(width=520, height=220, title="DatetimeTickFormatter", x_axis_type="datetime")
p_dt_fmt.line(dates, values, line_width=2)
p_dt_fmt.xaxis.formatter = DatetimeTickFormatter(days="%b %d", months="%b %Y")
show(p_dt_fmt)
```

```python
# LogTickFormatter
p_log_fmt = figure(width=520, height=220, title="LogTickFormatter", y_axis_type="log")
p_log_fmt.line([1, 2, 3, 4], [1, 10, 100, 1000], line_width=2)
p_log_fmt.yaxis.formatter = LogTickFormatter()
show(p_log_fmt)
```

```python
# MercatorTickFormatter
p_merc_fmt = figure(
    width=520,
    height=220,
    title="MercatorTickFormatter",
    x_axis_type="mercator",
    y_axis_type="mercator",
)
p_merc_fmt.line(x_merc, y_merc, line_width=2)
p_merc_fmt.xaxis.formatter = MercatorTickFormatter(dimension="lon")
p_merc_fmt.yaxis.formatter = MercatorTickFormatter(dimension="lat")
show(p_merc_fmt)
```

```python
# NumeralTickFormatter
p_num = figure(width=520, height=220, title="NumeralTickFormatter")
p_num.line([0, 1, 2, 3], [0.12, 0.21, 0.18, 0.31], line_width=2)
p_num.yaxis.formatter = NumeralTickFormatter(format="0.0%")
show(p_num)
```

```python
# PrintfTickFormatter
p_printf = figure(width=520, height=220, title="PrintfTickFormatter")
p_printf.line([0, 1, 2, 3], [3.4, 5.2, 2.8, 6.1], line_width=2)
p_printf.yaxis.formatter = PrintfTickFormatter(format="%.2f units")
show(p_printf)
```

```python
# TimedeltaTickFormatter
p_td_fmt = figure(width=520, height=220, title="TimedeltaTickFormatter", x_axis_type="timedelta")
p_td_fmt.line(dur_ms, metric, line_width=2)
p_td_fmt.xaxis.formatter = TimedeltaTickFormatter()
show(p_td_fmt)
```

`TickFormatter` is the abstract base class. For API design, model your options around concrete formatter classes and any class-specific kwargs (`format`, `precision`, `dimension`, etc.).

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.palettes import Category20

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y1=[1, 2, 4, 3, 4],
    y2=[1, 4, 2, 2, 3],
    y3=[2, 3, 4, 5, 6],
    y4=[3, 4, 3, 4, 5],
))
p = figure(width=400, height=400)

p.vline_stack(['y1', 'y2', 'y3', 'y4'], x='x', color=Category20[4], source=source)

show(p)
```
